### Dados:

- Rede de lojas: 19 lojas
- Abastecimento das lojas: 3 Centros Logísticos
- Objetivo: encontrar a posição ótima dos centros logísticos para distribuir entre as lojas
    - Centróides: Centros Logísticos
    - Formar clusters de lojas

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
dataset = np.array(
    #matriz com as coordenadas geográficas de cada loja
    [
     [-25, -46], #são paulo
     [-22, -43], #rio de janeiro
     [-25, -49], #curitiba
     [-30, -51], #porto alegre
     [-19, -43], #belo horizonte
     [-15, -47], #brasilia
     [-12, -38], #salvador
     [-8, -34], #recife
     [-16, -49], #goiania
     [-3, -60], #manaus
     [-22, -47], #campinas
     [-3, -38], #fortaleza
     [-21, -47], #ribeirão preto
     [-23, -51], #maringa
     [-27, -48], #florianópolis
     [-21, -43], #juiz de fora
     [-1, -48], #belém
     [-10, -67], #rio branco
     [-8, -63] #porto velho
    ]
     )
```

```python
x = dataset[:,1]
y = dataset[:,0]

plt.scatter(x, y)
plt.grid()
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, max_iter=100)

y_hat = kmeans.fit_predict(dataset)
```

```python
plt.scatter(dataset[:,1], dataset[:,0], c=y_hat)
plt.grid()

plt.scatter(kmeans.cluster_centers_[:,1], kmeans.cluster_centers_[:,0], lw=8.0, c='red', marker='s')
plt.show()

print('Coordenadas dos centróides:', kmeans.cluster_centers_[:,1], kmeans.cluster_centers_[:,0])
```
