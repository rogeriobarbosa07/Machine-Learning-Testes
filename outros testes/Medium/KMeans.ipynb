```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

### Dados de exemplo

```python
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.7, random_state=0)
```

visualizando as coordenadas de cada ponto

```python
# Criaremos alguns dados de exemplo para trabalhar. Imagine que são dados
# sobre clientes, com dois recursos, como "renda anual" e "pontuação de gastos".

plt.scatter(X[:,0], X[:,1], s=50)
plt.title('Dados dos clientes')
plt.xlabel('Renda anual')
plt.ylabel('Pontuação de gastos')
plt.grid()
plt.show()
```

### Criando o modelo K-Means


```python
kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(X)
y_hat = kmeans.predict(X)

# Nessa parte poderia usar a função fit_predict(X)
```

### Visualizando os Clusters

```python
plt.scatter(X[:, 0], X[:, 1], s=50, c=y_hat)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], c='red', lw=7.0, marker='s')
plt.grid()
plt.show()
```
