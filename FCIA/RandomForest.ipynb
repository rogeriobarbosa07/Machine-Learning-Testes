```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

```python
df = pd.read_csv("insurance.csv", keep_default_na=False)
df.head()
```

```python
df = df.drop(columns=['Unnamed: 0'])
df.head()
```

```python
y = df.iloc[:,7].values
X = df.drop(df.columns[7], axis=1).values
X
```

```python
labelencoder = LabelEncoder()
for i in range(X.shape[1]):
    if X[:,i].dtype == 'object':
        X[:,i] = labelencoder.fit_transform(X[:,i])
X
```

```python
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.3, random_state=12)
```

```python
# criando o modelo randomforest
modelo = RandomForestClassifier(random_state=1, max_depth=20, max_leaf_nodes=12, n_estimators=500)
modelo
```

```python
modelo.fit(X_treino, y_treino) # treinando o modelo randomforest
```

```python
tree_index = 1
tree_to_visualize = modelo.estimators_[tree_index]
plt.figure(figsize=(20,20))
plot_tree(tree_to_visualize, filled=True, feature_names=df.columns[:-1], class_names=True, rounded=True)
plt.show
```

```python
# chamando o modelo treinado para fazer as previsões
previsoes = modelo.predict(X_teste)
previsoes
```

```python
# calculando percentual de acerto, de acordo com os valores corretos e os previstos
accuracy = accuracy_score(y_teste, previsoes)
precision = precision_score(y_teste, previsoes, average='weighted')
recall = recall_score(y_teste, previsoes, average='weighted')
f1 = f1_score(y_teste, previsoes, average='weighted')

print(f'acurácia: {accuracy}\nprecisão: {precision}\nrecall: {recall}\nf1: {f1}')
```

```python
report = classification_report(y_teste, previsoes)
print(report)
```
