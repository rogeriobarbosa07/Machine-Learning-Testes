```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn import datasets
```

```python
iris = datasets.load_iris()
previsores = iris.data
classe = iris.target
previsores
```

### Dimensionamento de características

```python
sc = StandardScaler()
previsores = sc.fit_transform(previsores)
previsores
```

### Train-test split

```python
X_treino, X_teste, y_treino, y_teste = train_test_split(previsores, classe, test_size=0.3, random_state=123)
```

### Modelo

```python
rf = RandomForestClassifier(n_estimators=100, random_state=1234) # modelo e hiperparametros
rf.fit(X_treino, y_treino) # treinando o modelo com os dados de treino
```

```python
previsoes = rf.predict(X_teste) # fazendo previsões ^y com base nas variáveis X (independentes) de teste
taxa_acerto = accuracy_score(y_teste, previsoes)
taxa_acerto
```

```python
pca = PCA(n_components=3)
previsores = pca.fit_transform(previsores)
previsores
```

```python
X_treino, X_teste, y_treino, y_teste = train_test_split(previsores, classe, test_size=0.3, random_state=123)
```

```python
rf = RandomForestClassifier(n_estimators=100, random_state=1234) # modelo e hiperparametros
rf.fit(X_treino, y_treino) # treinando o modelo com os dados de treino
previsoes = rf.predict(X_teste) # fazendo previsões ^y com base nas variáveis X (independentes) de teste
taxa_acerto = accuracy_score(y_teste, previsoes)
taxa_acerto
```
