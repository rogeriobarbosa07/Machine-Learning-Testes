```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, ConfusionMatrixDisplay
from sklearn.datasets import load_iris
```

```python
dataset = load_iris()
dataset
```

```python
X_treino, X_teste, y_treino, y_teste = train_test_split(dataset.data, dataset.target, test_size=0.3)
```

```python
redeneural = MLPClassifier(
    hidden_layer_sizes=(4, 5),
    batch_size=20,
    activation='logistic',
    learning_rate='adaptive',
    max_iter=1300,
    verbose=True
)

redeneural.fit(X_treino, y_treino) # treinando
```

```python
plt.plot(redeneural.loss_curve_)
plt.title("Curva de erro (loss curve)")
plt.xlabel("Número de iterações")
plt.ylabel("Valor do erro")
```

```python
y_hat = redeneural.predict(X_teste) # fazendo previsões (a partir dos valores de X)
print("acurácia =", accuracy_score(y_teste, y_hat)) # comparando classes reais (y) com previsões (y^)
```

```python
cm = confusion_matrix(y_teste, y_hat)
ConfusionMatrixDisplay(confusion_matrix=cm).plot()
plt.title("Matriz de Confusão")
plt.ylabel("Classes reais")
plt.xlabel("Classes previstas")
```
