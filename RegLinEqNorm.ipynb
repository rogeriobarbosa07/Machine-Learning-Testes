```python
# Exemplo: Salário de funcionários por meses trabalhados

import numpy as np
import matplotlib.pyplot as plt

x = np.array([3.0, 6.0, 9.0, 12.0, 18.0, 24.0, 30.0, 36.0])
y = np.array([2000.00, 2300.00, 2600.00, 3000.00, 3500.00, 4000.00, 4200.00, 4500.00])

media_x = np.mean(x)
media_y = np.mean(y)
media_xy = np.mean(x*y)
media_x2 = np.mean(x**2)
```

```python
theta1 = (media_xy - media_x*media_y)/(media_x2 - (media_x)**2)
theta0 = media_y - (theta1*media_x)
print("valor de theta 1:", theta1)
print("valor de theta 0:", theta0)
```

```python
novo_x = float(input("Insira seus meses de trabalho: "))
previsao = theta0 + (theta1 * novo_x)
print("Sua previsão salarial: ", previsao)
```

```python
# Visualizando as previsões
salario = theta0+theta1*x
plt.scatter(x, y, label='Dados reais', color='blue')
plt.plot(x, salario, color='red', label='Regressão Linear')
plt.xlabel('Meses de trabalho (x)')
plt.ylabel('Salário (y)')
plt.show()
```
